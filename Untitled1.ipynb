```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
#Condições e variáveis para a análise dos dados e plotagem dos gráficos
ano = [2014, 2017, 2018]
dados2014 = 0
dados2017 = 0
dados2018 = 0
coluna1 = 2
coluna2 = 5
separador = ";"

for y in ano:
    if y > 2014: 
        coluna1 = 3
        coluna2 = 6
        separador = ","

#Leitura das colunas e linha que serão analisadas
    a=pd.read_csv(f"MICRODADOS_ENCCEJA_NACIONAL_REGULAR_{y}.csv",encoding="ANSI",sep=separador, usecols=[coluna1, coluna2], nrows=20001)

#Lista de todos os Estados
    estados = {"SP":0, 'GO':0, 'TO':0, 'RS':0, 'DF':0, 'BA':0, 'MS':0, 'PE':0, 'RJ':0, 'SC':0, 'MG':0,
       'PR':0, 'MT':0, 'AL':0, 'PA':0, 'CE':0, 'AC':0, 'ES':0, 'RN':0, 'PB':0, 'RR':0, 'AM':0,
       'PI':0, 'MA':0, 'RO':0, 'SE':0, 'AP':0}

#Fórmula para tirar a média das idades dos alunos de cada região
    for estado in estados:
        estados[estado] = a.query (f"SG_UF_RESIDENCIA=='{estado}'")["NU_IDADE"].sum()//a.query (f"SG_UF_RESIDENCIA=='{estado}'")["NU_IDADE"].count()
    if y == 2014:
        dados2014 = estados
    if y == 2017:
        dados2017 = estados
    if y == 2018:
        dados2018 = estados
```

```python
#Função para plotar o gráfico de barras horizontal
plt.rcParams['figure.figsize'] = (20, 50)

#Plot do gráfico de 2014
for x in dados2014:
    plt.barh(x, dados2014[x])
        
```

```python
#Plot do gráfico de 2017
for x in dados2017:
    plt.barh(x, dados2017[x])
        
```

```python
#Plot do gráfico de 2018
for x in dados2018:
    plt.barh(x, dados2018[x])
```
